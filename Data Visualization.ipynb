```python
import os
import random
import matplotlib.pyplot as plt
import cv2
import seaborn as sns
```

```python
def display_image_grid(images, labels, ncols=5):
    n_images = len(images)
    nrows = n_images // ncols + (n_images % ncols > 0)
    
    plt.figure(figsize=(15, 3 * nrows))
    
    for i in range(n_images):
        plt.subplot(nrows, ncols, i + 1)
        plt.imshow(images[i])
        plt.title(labels[i])
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Load a few sample images and labels for grid display
sample_images = []
sample_labels = []

for label in os.listdir("D:\\movie-recommendation-system\\Emotion_detection\\data\\train"):
    label_folder = os.path.join("D:\\movie-recommendation-system\\Emotion_detection\\data\\train", label)
    for img_name in random.sample(os.listdir(label_folder), 5):  # Take 5 samples from each class
        img_path = os.path.join(label_folder, img_name)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        sample_images.append(img)
        sample_labels.append(label)

display_image_grid(sample_images, sample_labels)
```

```python
def plot_class_distribution(folder):
    classes = os.listdir(folder)
    counts = [len(os.listdir(os.path.join(folder, label))) for label in classes]
    
    plt.figure(figsize=(10, 5))
    sns.barplot(x=classes, y=counts)
    plt.xlabel("Emotion Classes")
    plt.ylabel("Number of Images")
    plt.title("Class Distribution in Training Dataset")
    plt.xticks(rotation=45)
    plt.show()

# Plot class distribution for training and testing datasets
plot_class_distribution("D:\\movie-recommendation-system\\Emotion_detection\\data\\train")
```
